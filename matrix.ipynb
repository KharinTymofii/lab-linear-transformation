# Перша частина

```python
def plot_object(points, title=''):
    plt.figure() 
    plt.plot(points[:, 0], points[:, 1])
    plt.title(title)
    plt.grid(True)
    plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

object_1 = np.array([[1, 1], [2, 4], [3, 2], [4, 5], [5, 3], [1, 1]])
object_2 = np.array([[2, 2], [3, 6], [5, 4], [7, 7], [6, 3], [2, 2]])

print(f'Object 1 \n {object_1}')
print(f'Object 2 \n {object_2}')
```

```python
plot_object(object_1, 'Object 1')
plot_object(object_2, 'Object 2')
```

### Обертання об'єкта на певний кут ###

```python
def rotation(points, angle):
    radian = np.radians(angle)
    rotation_matrix = np.array([
        [np.cos(radian), -np.sin(radian)],
        [np.sin(radian), np.cos(radian)]
    ])
    return np.dot(points, rotation_matrix.T) # матриця повороту транспонована, щою повертати проти годиникової стрілки

# Вікіпедія про поворот матриці: https://en.wikipedia.org/wiki/Rotation_matrix
```

```python
rotated_object_1 = rotation(object_1, 270)
print(f'Firts matrix \n {rotated_object_1}')
plot_object(rotated_object_1, 'Object 1 rotated')

rotated_object_2 = rotation(object_2, 48)
print(f'Second matrix \n {rotated_object_2}')
plot_object(rotated_object_2, 'Object 2 rotated')
```

### Маштабування об'єкта ###

```python
def scaling(points, coefficient_x, coefficient_y=1):
    scaling_matrix = np.array([
        [coefficient_x, 0], 
        [0, coefficient_y]])
    return np.dot(points, scaling_matrix)
```

```python
object_1_scaling = scaling(object_1, 20, 7)
print(f'Firts matrix \n {object_1_scaling}')
plot_object(object_1_scaling, 'Object 1 scaling')

object_2_scaling = scaling(object_2, 4, 8)
print(f'Second matrix \n {object_2_scaling}')
plot_object(object_2_scaling, 'Object 2 scaling')
```

### Віддзеркалення відно осі ###

```python
def reflection(points, axis):
    if axis == 'y':
        reflection_matrix = np.array([
            [-1, 0], 
            [0, 1]])
    elif axis == 'x':
        reflection_matrix = np.array([
            [1, 0], 
            [0, -1]])
    return np.dot(points, reflection_matrix)
```

```python
object_1_refletion = reflection(object_1, 'x')
print(f'Firts matrix \n {object_1_refletion}')
plot_object(object_1_refletion, 'Reflected object 1 X')

object_2_refletion = reflection(object_2, 'x')
print(f'Second matrix \n {object_2_refletion}')
plot_object(object_2_refletion, 'Reflected object 2 X')

object_1_refletion = reflection(object_1, 'y')
print(f'Third matrix \n {object_1_refletion}')
plot_object(object_1_refletion, 'Reflected object 1 Y')

object_2_refletion = reflection(object_2, 'y')
print(f'Fourth matrix \n {object_2_refletion}')
plot_object(object_2_refletion, 'Reflected object 2 Y')
```

### Нахил об'єкта

```python
def shear(points, x=0, y=0):
    shear_matrix = np.array([
        [1, np.tan(x)], 
        [np.tan(y), 1]])
    return np.dot(points, shear_matrix.T)
```

```python
object_1_sheared_x = shear(object_1, 0.9, 0)
print(f'Firts matrix \n {object_1_sheared_x}')
plot_object(object_1_sheared_x, 'Object 1 sloped on X')

object_1_sheared_y = shear(object_1, 0, 1.3)
print(f'Second matrix \n {object_1_sheared_y}')
plot_object(object_1_sheared_y, 'Object 1 sloped on Y')

object_1_sheared_xy = shear(object_1, 0.9, 1.3)
print(f'Third matrix \n {object_1_sheared_xy}')
plot_object(object_1_sheared_xy, 'Object 1 sloped on X & Y')

object_2_sheared_x = shear(object_2, 0.9, 0)
print(f'Fourth matrix \n {object_2_sheared_x}')
plot_object(object_2_sheared_x, 'Object 1 sloped on X')

object_2_sheared_y = shear(object_2, 0, 1.3)
print(f'Fifth matrix \n {object_2_sheared_y}')
plot_object(object_2_sheared_y, 'Object 1 sloped on Y')

object_2_sheared_xy = shear(object_2, 0.9, 1.3)
print(f'Sixth matrix \n {object_2_sheared_xy}')
plot_object(object_2_sheared_xy, 'Object 1 sloped on X & Y')
```

### Трансформація

```python
def transformation(points, matrix):
    return np.dot(points, matrix.T)
```

```python
custom_matrix = np.array([
    [1, 4],
    [6, 8]
])

object_1_transformation = transformation(object_1, custom_matrix)
print(f'Firts matrix \n {object_1_transformation}')
plot_object(object_1_transformation, 'Object 1 transformation')

object_2_transformation = transformation(object_2, custom_matrix)
print(f'Firts matrix \n {object_2_transformation}')
plot_object(object_2_transformation, 'Object 2 transformation')
```

### 3Д ОБ'ЄКТИ

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_object(points, title=''):
    figure = plt.figure()
    ax = figure.add_subplot(projection='3d')
    ax.plot(points[:, 0], points[:, 1], points[:, 2])
    ax.set_title(title)
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    plt.show()
```

```python
object_3d = np.array([[0, 0, 0], [1, 2, 0], [2, 1, 3], [3, 0, 1], [0, 0, 0]])
print(object_3d)
plot_3d_object(object_3d)
```

#### Обертання 3д об'єкта

```python
def rotation_3d(points, angel, axis='x'):
    radian = np.radians(angel)
    if axis == 'x':
        rotation_matrix = np.array([
            [1, 0, 0],
            [0, np.cos(radian), -np.sin(radian)],
            [0, np.sin(radian), np.cos(radian)]])
    elif axis == 'y':
        rotation_matrix = np.array([
            [np.cos(radian), 0, np.sin(radian)],
            [0, 1, 0],
            [-np.sin(radian), 0, np.cos(radian)]])
    elif axis == 'z':  
        rotation_matrix = np.array([
            [np.cos(radian), -np.sin(radian), 0],
            [np.sin(radian), np.cos(radian), 0],
            [0, 0, 1]])
    return np.dot(points, rotation_matrix)
    
## інформацію про обертання матриць брав з https://en.wikipedia.org/wiki/Rotation_matrix    
```

```python
rotated_object_3d_x = rotation_3d(object_3d, 40, 'x')
print(f'Firts matrix \n {rotated_object_3d_x}')
plot_3d_object(rotated_object_3d_x)

rotated_object_3d_y = rotation_3d(object_3d, 130, 'y')
print(f'Second matrix \n {rotated_object_3d_y}')
plot_3d_object(rotated_object_3d_y)

rotated_object_3d_z = rotation_3d(object_3d, 70, 'z')
print(f'Third matrix \n {rotated_object_3d_z}')
plot_3d_object(rotated_object_3d_z)
```

#### Маштабування 3д об'єктів

```python
def scaling_3d(points, x, y, z):
    scaling_matrix = np.array([
        [x, 0, 0],
        [0, y, 0],
        [0, 0, z]])
    return np.dot(points, scaling_matrix)
```

```python
scaled_object_3d = scaling_3d(object_3d, 29, 9, 11)
print(f'Firts matrix \n {scaled_object_3d}')
plot_3d_object(scaled_object_3d)
```

### Зображення

```python
def show_image(image, title=''):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
import cv2 as cv

image = cv.imread('kse.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
show_image(image)
```

#### Обертання

```python
def rotation_image(rows, cols, angel, zoom):
    M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0), angel, zoom)
    dst = cv.warpAffine(image, M, (cols,rows))                
    show_image(dst)
```

```python
rows,cols = image.shape[:2]
rotation_image(rows, cols, 130, 0.7)
```

#### Маштабування зображення 

```python
def scale_image(image, x, y):
    scaled = cv.resize(image, None, fx=x, fy=y)
    show_image(scaled)
```

```python
scale_image(image, 5, 2)
```

#### Віддзеркалювання зображення

```python
def mirror_image(image, axis):
    if axis == 'x':
        mirrored = cv.flip(image, 0)
    elif axis == 'y':
        mirrored = cv.flip(image, 1)
    show_image(mirrored)
```

```python
mirror_image(image, 'y')
```
